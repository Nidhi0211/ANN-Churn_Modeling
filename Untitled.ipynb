```python
import numpy as np
import matplotlib.pyplot as plot
import pandas as mypd
import GradientDescent
```

```python
myData=mypd.read_csv(".\data.csv",header=None)
myData
```

```python
def plot_points(X,Y):
    admitted=X[np.argwhere(Y==1)]
    rejected=X[np.argwhere(Y==0)]
    plot.scatter([s[0][0] for s in rejected],[s[0][1] for s in rejected],s=30,color='blue',edgecolor='k')
    plot.scatter([s[0][0] for s in admitted],[s[0][1] for s in admitted],s=30,color='red',edgecolor='k')
```

```python
X=np.array(myData[[0,1]])
X
                
```

```python
Y=np.array(myData[[2]])
Y
```

```python
plot_points(X,Y)
plot.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def output_formula(features,weight,bias):
    return sigmoid(np.dot(features,weight)+bias)

def error_for(y,yp):
    return -y*np.log(yp)-(1-y)*np.log(1-yp)

def update_w(X,Y,weights,bias,learnrate):
    output=output_formula(X,weights,bias)
    d_err=-(Y-output)
    weights -= learnrate * d_err * X
    bias -= learnrate * d_err
    return weights, bias
    
```

```python
epochs=100
GradientDescent.train(X,Y,epochs,learnrate=0.01,sigmoid=sigmoid,output_formula=output_formula,
                      error_formula=error_for,update_weights=update_w,graph_lines=True)
```
